```python
import pandas as pd
import io

df = pd.read_csv('/content/drive/MyDrive/Machine Learning/wbc.csv')
```

# 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
df_use = df.drop(['id'], axis=1)
df_use = df_use.drop(['Unnamed: 32'], axis=1)

df_use.head()

# df_not_use = df[['id']]
# df_not_use.head()
```

# 2. Lakukan proses encoding pada kolom "diagnosis".

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

le = LabelEncoder()
df_use['diagnosis'] = le.fit_transform(df_use['diagnosis'])
df_use.head()
```

# 3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()
numeric = df_use.columns.difference(['diagnosis'])
df_use[numeric] = std.fit_transform(df_use[numeric])
df_use.head()
```

# 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
from sklearn.model_selection import train_test_split

df_wbs_train, df_wbs_test = train_test_split(df_use, test_size=0.2, random_state=0, stratify=df_use['diagnosis'])

print(f'Jumlah label data asli:\n{df_use.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_wbs_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_wbs_test.diagnosis.value_counts()}')
```
